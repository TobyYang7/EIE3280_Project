```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import scienceplots as sp
plt.style.use('science')
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

unsplash_data = pd.read_csv("data/unsplash.csv")
pexels_data = pd.read_csv("data/pexels.csv")
px_data = pd.read_csv("data/500px.csv")
tuchong_data = pd.read_csv("data/tuchong.csv")

# Convert the date column to datetime
unsplash_data['Unnamed: 0'] = pd.to_datetime(
    unsplash_data['Unnamed: 0'], format='%Y%m%d')
pexels_data['Unnamed: 0'] = pd.to_datetime(pexels_data['Unnamed: 0'], format='%Y%m%d')
px_data['Unnamed: 0'] = pd.to_datetime(px_data['Unnamed: 0'], format='%Y%m%d')
tuchong_data['Unnamed: 0'] = pd.to_datetime(tuchong_data['Unnamed: 0'], format='%Y%m%d')

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(unsplash_data['Unnamed: 0'],
         unsplash_data['Desktop'], label='Pexels')
plt.plot(pexels_data['Unnamed: 0'], pexels_data['Desktop'],label='Unsplash')
plt.plot(px_data['Unnamed: 0'], px_data['Desktop'],label='500px')
plt.plot(tuchong_data['Unnamed: 0'], tuchong_data['Desktop'],label='Tuchong')
plt.xlabel('Date', fontsize=20)
plt.ylabel('Number of Devices', fontsize=20)
plt.title('All Devices Usage Over Time', fontsize=24)
plt.grid(True)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=7))
plt.gcf().autofmt_xdate()  # Rotate date labels
plt.legend()
plt.tight_layout()

plt.show()
```
